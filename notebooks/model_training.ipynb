```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import KFold, cross_val_score
```

```python
data_path="C:\\Users\\Mahakaal\\Documents\\HOUSE-PRICE-PREDICTION\\data\\proccessed\\bengaluru_vectorized_final.csv"
df = pd.read_csv(data_path)
```

```python
df['bin_cut'] = pd.qcut(df['price'],q=5, labels=False)

x=df.drop(columns=['price'])
y=df['price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,stratify=df['bin_cut'], random_state=1)
x_train, x_test, y_train, y_test = x_train.drop(columns=['bin_cut']), x_test.drop(columns=['bin_cut']), y_train.drop(columns=['bin_cut']), y_test.drop(columns=['bin_cut'])

lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
y_pred = lr_model.predict(x_test)

```

```python
y_pred = np.clip(y_pred, a_min=0, a_max=None)
```

```python
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"R-squared (R²) Score: {r2:.4f}")
print(f"Average Prediction Error (RMSE): {rmse:.2f} Lakhs")

kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(lr_model, x, y, cv=kf, scoring="r2")

print("  SHUFFLED 5-FOLD CROSS-VAL   ")
print(f"All Shuffled CV R² Scores: {cv_scores}")
print(f"Mean R² Score: {cv_scores.mean():.4f}")
print(f"Standard Deviation: {cv_scores.std():.4f}")
```
